```python
import numpy as np
from matplotlib import pyplot as plt
data  = np.loadtxt("../mpc_wavefront_iter_0.txt", delimiter=' ')
```

```python
np.unique(data[:, -1])
```

```python
keys = [0, 1, -2, -3, -4]
colors = ["yellow", "blue", "black", "green", "red"]
keys_to_colors = {}
fig = plt.figure(figsize=(20, 20))

ax = fig.add_subplot(111)
for key, color in zip(keys, colors):
    keys_to_colors[key] = color
    ax.scatter(data[data[:, -1] == float(key), 0], data[data[:, -1] == float(key), 1], color=color, marker='o', s=2)
    # else:
    #     plt.scatter(data[data[:, -1] == float(key), 0], data[data[:, -1] == float(key), 1], color=color, marker='o', s=0.1)
    
plt.show()
```
